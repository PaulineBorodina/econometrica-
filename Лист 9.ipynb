```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import re
import scipy.stats as stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

$$ H_0: \beta{south} = \beta{totwrk*south} = \beta{age*south} = \beta{smsa*south} = 0 \\ 
$$$$ H_1: \beta{south}^2 + \beta{totwrk*south}^2 + \beta{age*south}^2 + \beta{smsa*south}^2 >0 $$

```python
date_string = '''R2                      0.115          0.131  '''
rest, unrest =np.float_(re.findall(r'([-+]?\d+.\d+)', date_string)) 
```

```python
F_obs = (unrest - rest)/(1-unrest)*(706-8)/4  
F_obs.round(2)
```

```python
alpha = 0.1 
F_cr = stats.f.ppf(1-alpha,4,706-8)
F_cr.round(2)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/diamonds.csv')
df
```

```python
df_all = pd.get_dummies(df.loc[:,['price','carat','color','cut','clarity','x','y','z']])
df_all.rename(columns = {'cut_Very Good':'cut_Very'}, inplace = True )
df_all
```

```python
model_d = smf.ols(formula='np.log(price)~carat+I(carat**2)+color_E+color_F+color_G+color_H+color_I+color_J+cut_Good+cut_Very+cut_Premium+cut_Ideal+x+y+z',data=df_all).fit()
model_d.params
```

$$ H_0: \beta{colorE} = \beta{colorF} = \beta{colorG} = \beta{colorH} = \beta{colorI} = \beta{colorJ} = 0 \\ $$

```python
model_d.f_test('color_E=color_F=color_G=color_H=color_I=color_J=0')
```

```python
stats.f.ppf(1-0.05,6,len(df_all)-6).round(2)
```
